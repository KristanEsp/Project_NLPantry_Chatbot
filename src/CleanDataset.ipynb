```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv("./archive/recipes_w_search_terms.csv")
```

```python
df.shape
```

```python
#Only take a number of recipe
recipe_total = 10000
df_random = df.sample(n = recipe_total)
#drop the ID column
df_random.drop(columns = "id", inplace = True)
df_random.head(10)
```

```python
#Sve the dataset into a csv file
df_random.to_csv("Recipe_Dataset.csv", index = False)
```

## Extract the ingredient list

```python
ingredient_list = []
for ingredient in df_random["ingredients"]:
    #Remove punctuations except for comma
    punctuations_with_comma = r"[''\[\]]"
    ingredient = re.sub(punctuations_with_comma, "", ingredient)
    #Put the detected ingredients into a list
    ingredient_list.extend(ingredient.split(', '))

#Remove duplicates in ingredient list
ingredient_list = list(set(ingredient_list))
ingredient_list

#Export the list into a txt file
with open("ingredients.txt", "w") as output:
    output.write(", ".join(ingredient_list))
```
